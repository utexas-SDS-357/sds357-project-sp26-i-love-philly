```python
# convert to datetime cols
eagles["FullDate"] = pd.to_datetime(eagles["FullDate"])
data["date"] = pd.to_datetime(data["date"])
```

```python
mask = data["date"].isin(eagles["FullDate"].unique())
eagles_gamedays = data[mask]
eagles_gamedays
```

```python
eagles_home_games = eagles[eagles["Home"] == 1]
eagles_home_games
```

```python
mask2 = data["date"].isin(eagles_home_games["FullDate"].unique())
eagles_home_gamedays = data[mask2]
eagles_home_gamedays
```

```python
eagles_home_gamedays.to_csv("STOPS_EAGLES_HOME_GAMES.csv", index=False)
```

```python
def add_game_indicator_col(stop_data, game_dates, new_col_name):

    # game_dates: unique dates

    # convert col to date
    stop_data["date"] = pd.to_datetime(stop_data["date"])
    mask = stop_data["date"].isin(game_dates)
    stop_data[new_col_name] = mask

    return stop_data
```
