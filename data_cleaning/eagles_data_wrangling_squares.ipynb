## Create square regions

### Drop unnecessary columns

```python
philly_stops = philly_stops.iloc[:, [1, 2, 4, 5]]
philly_stops.columns
```

### Remove NA value rows

```python
# Check % of na values first
philly_stops.isna().mean() * 100

philly_stops = philly_stops.dropna()
philly_stops
```

### Convert latitude and longitude to Km based coordinates and create 1 km square grid

```python
km_to_deg_lat = 1 / 111
mean_lat = philly_stops['lat'].mean()

# Longitude conversion depends on latitude
km_to_deg_lng = 1 / (111 * np.cos(np.radians(mean_lat)))

# Bin each point the floor km
philly_stops['lat_bin'] = np.floor(philly_stops['lat'] / km_to_deg_lat)
philly_stops['lng_bin'] = np.floor(philly_stops['lng'] / km_to_deg_lng)

# Create a region ID
philly_stops['region'] = philly_stops['lat_bin'].astype(int).astype(str) + "_" + philly_stops['lng_bin'].astype(int).astype(str)

print(philly_stops[['lat', 'lng', 'region']].head())
```

```python
print(km_to_deg_lng)
```

### Plot grid on a map

```python
# Center map
center_lat = philly_stops['lat'].mean()
center_lng = philly_stops['lng'].mean()

m = folium.Map(location=[center_lat, center_lng], zoom_start=11)

# Get unique grid cells
unique_cells = philly_stops[['lat_bin', 'lng_bin']].drop_duplicates()

for _, row in unique_cells.iterrows():
    lat_bin = row['lat_bin']
    lng_bin = row['lng_bin']

    # Convert back to coordinates (bottom-left corner)
    lat_min = lat_bin * km_to_deg_lat
    lng_min = lng_bin * km_to_deg_lng

    lat_max = lat_min + km_to_deg_lat
    lng_max = lng_min + km_to_deg_lng

    # Draw rectangle
    folium.Rectangle(
        bounds=[[lat_min, lng_min], [lat_max, lng_max]],
        color='blue',
        weight=1,
        fill=False
    ).add_to(m)

# Optional: plot some points
# for _, row in philly_stops.sample(10000).iterrows():  # sample for performance
#     folium.CircleMarker(
#         location=[row['lat'], row['lng']],
#         radius=.2,
#         color='red',
#         fill=True
#     ).add_to(m)

m
```

## Aggregate stops by date, region, and hour

### Convert time to hour block

```python
philly_stops['hour'] = pd.to_datetime(philly_stops['time'], format='%H:%M:%S').dt.hour
philly_stops.head()
```

```python
# Drop time column
philly_stops.drop(columns=['time'], inplace=True)
philly_stops.head()
```

### Aggregate (count) the number of stops for each date, region, hour combination

```python
aggregate_philly_stops = (
    philly_stops
    .groupby(['date', 'hour', 'region'])
    .size()
    .reset_index(name='stop_count')
)

aggregate_philly_stops
```

## Fill 0 stops for any missing combinations

```python
# Create full date range
all_dates = pd.date_range(start=aggregate_philly_stops['date'].min(), end=aggregate_philly_stops['date'].max(), freq='D')

# Define full hour range (0-23)
all_hours = range(0, 24)

# Get all regions
all_regions = aggregate_philly_stops['region'].unique()

# Create full combination grid
full_index = pd.MultiIndex.from_product(
    [all_dates, all_hours, all_regions],
    names=['date', 'hour', 'region']
)

# Reindex
aggregate_philly_stops = (
    aggregate_philly_stops.set_index(['date', 'hour', 'region'])
      .reindex(full_index)
      .reset_index()
)

# Fill missing values with 0 stops
aggregate_philly_stops['stop_count'] = aggregate_philly_stops['stop_count'].fillna(0).astype(int)
aggregate_philly_stops['date'] = pd.to_datetime(aggregate_philly_stops['date']).dt.date
aggregate_philly_stops
```

## Add game information and home game flag to stop data

### First, check that all gamedays are in the stop data

 ### Left join philly stops with Eagles data

```python
# Merge (left join keeps all philly_stops rows)
# non-gamedays will have NA in all Eagles columns
merged_df = aggregate_philly_stops.merge(
    eagles,
    how='left',
    left_on='date',
    right_on='Date'
)
```

### Create gameday indicator column

```python
# Create gameday indicator (1 if matched with Eagles game, else 0)
merged_df['gameday'] = merged_df['Date'].notna().astype(int)

# Drop the Eagles date column
merged_df = merged_df.drop(columns=['Date'])
merged_df
```

## Add before/after column for gamedays

### Filter to just gamedays

```python
gamedays = merged_df[merged_df['gameday'] == 1].copy()
gamedays
```

### Create start/end columns that are rounded to the nearest hour

```python
gamedays['start_hour'] = (pd.to_datetime(gamedays['StartTime'], format='%H:%M:%S')
                          .dt.round('h') # round to the nearest hour
                          .dt.hour
                          )
gamedays['end_hour'] = (pd.to_datetime(gamedays['xEndTime'], format='%H:%M:%S')
                        .dt.round('h') # round to the nearest hour
                        .dt.hour
                        )
gamedays
```

### Classify each row as before/after/during the game

```python
conditions = [
    gamedays['hour'] < gamedays['start_hour'],
    (gamedays['hour'] >= gamedays['start_hour']) & (gamedays['hour'] <= gamedays['end_hour']),
    gamedays['hour'] > gamedays['end_hour']
]

choices = ['before', 'during', 'after']

gamedays['game_period'] = np.select(conditions, choices, default="unknown")
gamedays
```
