# Cleaning Eagles Data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/home/jasmine/PROJECTS/sds357-project-sp26-i-love-philly/data/Eagles Schedule & Scores - Eagles.csv", 
                   header=1)
data.rename(columns={'Unnamed: 3': 'Time', 'Unnamed: 5': "Result", "Unnamed: 8": "H/A",
                     "Tm": "TeamScore", "Opp.1": "OppScore"}, inplace=True)
data
```

```python
data.dtypes
```

```python
# drop rows where week is NaN (spacers)
data = data.dropna(subset=["Week"])
# drop bye weeks
data = data[data["Opp"] != "Bye Week"]
data
```

```python
# add point differential col (eagles - opponent)Eagles\ Schedule\ \&\ Scores\ -\ Eagles.csv
data["Differential"] = data["TeamScore"] - data["OppScore"]
data
```

```python
# home/away col
data["Home"] = np.where(data["H/A"] == "@", 1, 0)
# W/L binary col
data["W"] = np.where(data["Result"] == "W", 1, 0)
data
```

```python
# get the full date
data["FullDate"] = data["Date"] + ", " + data["Year"].astype(str)
data["FullDate"] = pd.to_datetime(data["FullDate"])
data
```

```python
time_str = data["Time"].str.split().str[0]

data["FullTime"] = pd.to_datetime(time_str,
                                  format='%I:%M%p',
                                  errors='coerce').dt.strftime('%H:%M:%S')
data
```

Other considerations: 
* encode opponents (one-hot) to see if certain opponents incite fans more (Cowboys, Giants mostly)
* encode game type (regular season vs postseason)
* encode day of week

```python
data.columns
```

```python
cols = ["Season", "Year", "Week", "Day", "FullDate", "FullTime", "Home", "Opp", "W", 
        "TeamScore", "OppScore", "Differential"]
data = data[cols]
data.to_csv("EAGLES_SCHEDULE_CLEANED.csv", index=False)
```
